```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
```

```python
sns.set(font_scale = 1.3)
np.random.seed(42)
np.set_printoptions(precision = 6, suppress = True)
```

```python
data, target = make_regression(n_samples = 200, n_features = 1, noise = 20)
target = target ** 2
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(max_depth = 3, criterion = "squared_error")
dtr.fit(data, target)
pred = dtr.predict(data)
```

```python
plot_data = np.linspace(-3, 3, 1000).reshape(-1, 1)
```

```python
fig = plt.figure(figsize = (10, 8))
plt.scatter(data, target)
plt.plot(plot_data, dtr.predict(plot_data), c = "darkgreen")
```

```python
from sklearn.metrics import mean_squared_error as mae
max_depth = [1, 4, 10, 20]
i = 1
fig = plt.figure(figsize = (30, 15))
for depth in max_depth:
  dtr = DecisionTreeRegressor(max_depth = depth)
  dtr.fit(data, target)
  plt.subplot(1, 4, i)
  plt.plot(plot_data, dtr.predict(plot_data), c = "green")
  plt.scatter(data, target, c = "red")
  plt.title("MAE: " +  str(mae(target, dtr.predict(data))))
  i += 1
```
